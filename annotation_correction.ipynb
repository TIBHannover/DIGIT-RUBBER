```python
with open('./annotation_v1.txt', 'r') as f:
    with open('./annotation_v3.ttl', 'w') as fw:
        for i, line in enumerate(f.readlines()):
            if re.findall('> <http.[^\s]*"$', line):
                line = re.sub('\"\n', "> .\n", line)
                # print(line)
            if re.findall('.*"\n$', line):
                line = re.sub('\"\n', "\" .\n", line)
                # print("#########################################", line)
            # print(line)
            fw.write(line)
```

```python
with open('./annotation_v3.ttl', 'r') as f:
    with open('./annotation_v4.ttl', 'w') as fw:
        for i, line in enumerate(f.readlines()):
            if re.findall('\"\"', line):
                line = re.sub('\"\"', "", line)
                # print(line)
            fw.write(line)
```

```python
with open('./annotation_v5.ttl', 'r') as f:
    with open('./annotation_v6.ttl', 'w') as fw:
        for i, line in enumerate(f.readlines()):
            if re.findall(',1\n', line):
                line = re.sub('\",1\n', "> 1 .\n", line)
                # print(line)
            fw.write(line)
```

```python
with open('./annotation_v6.ttl', 'r') as f:
    with open('./annotation_v7.ttl', 'w') as fw:
        for i, line in enumerate(f.readlines()):
            if re.findall('1.1\/date\"', line):
                line = re.sub('1.1/date\",', '1.1/date> ', line)
                line = re.sub('\n', ' .\n', line)
                print(line)
            fw.write(line)
```

```python
with open('./annotation_v7.ttl', 'r') as f:
    with open('./annotation_v8.ttl', 'w') as fw:
        for i, line in enumerate(f.readlines()):
            if not re.findall("^[<http:]", line):
                if not re.findall("\" .\n", line):
                    line = re.sub('\n', ' ', line)
                    print(line)
            fw.write(line)
```

```python
# Find relevant annotation properties in file

relevant_ann_l = []
with open('./WebProtege/digitrubber-full.ttl', 'r') as f:
    for line in f.readlines():
        if ' rdf:type owl:AnnotationProperty' in line:
            relevant_ann_l.append(line.split(sep=' rdf:type owl:AnnotationProperty')[0])
```
